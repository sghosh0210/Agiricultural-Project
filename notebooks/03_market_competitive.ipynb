```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/raw/india_agriculture_seed_sales_data.csv')
```

Competitive Intensity by Region

```python
comp_region = df.groupby('Region').agg(
    Avg_Competitor_Share=('Competitor_Market_Share_%','mean'),
    Avg_Profit_Margin=('Profit_Margin_%','mean'),
    Avg_Farmer_Sentiment=('Farmer_Sentiment_Score','mean')
).round(2)
print(comp_region)
```

Price Gap Analysis

```python
df['Price_Ratio_Wholesale'] = df['Seed_Price_Per_Unit'] / df['Wholesale_Price']
df['Price_Ratio_Retail'] = df['Seed_Price_Per_Unit'] / df['Retail_Price']

price_analysis = df.groupby('Vegetable_Type').agg(
    Avg_Seed_Price_per_unit=('Seed_Price_Per_Unit','mean'),
    Avg_Wholesale=('Wholesale_Price','mean'),
    Avg_Retail=('Retail_Price','mean'),
    Profit_Margin=('Profit_Margin_%','mean')
).round(2)
print(price_analysis)
```

```python
from scipy.stats import pearsonr

corr, p = pearsonr(df['Competitor_Market_Share_%'], df['Profit_Margin_%'])
print(f"Correlation (Competitor Share vs Profit): {corr:.3f}, p={p:.4f}")
# Expected: negative — higher competitor share → lower our margin
```

We found that competitor market share doesn't affect our profit margins. Whether competitors have 20% or 40% market share, our profits stay the same. This means we're not competing on price - we have our own loyal customer base
